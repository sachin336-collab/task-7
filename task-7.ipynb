```python

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("chiku_data.db")

```

```python
query = """
SELECT 
    product, 
    SUM(quantity) AS total_qty, 
    SUM(quantity * price) AS revenue 
FROM sales 
GROUP BY product
"""
```

```python
df = pd.read_sql_query(query, conn)
```

```python
conn.close()
```

```python
print("ðŸ“Š Sales Summary:\n")
print(df)
```

```python
df.plot(kind='bar', x='product', y='revenue', color='skyblue', legend=False)
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue (â‚¹)")
plt.tight_layout()
```

```python
plt.savefig("sales_chart.png")
plt.show()
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("chiku_data.db")

query = """
SELECT 
    product, 
    SUM(quantity * price) AS revenue 
FROM sales 
GROUP BY product 
ORDER BY revenue DESC 
LIMIT 3
"""

df = pd.read_sql_query(query, conn)
conn.close()

print(df)
```
